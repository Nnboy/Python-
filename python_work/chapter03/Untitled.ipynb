```python
unconfirmed_user = list(map(str,range(1,5)))
```

```python
confirmed_users = []

while unconfirmed_user :
    current_user = unconfirmed_user.pop()
    
    print('Verfying user: ' + current_user.title())
    confirmed_users.append(current_user)

print("\nThe following users have been condirmed:")
for confirmed_user in confirmed_users:
    print(confirmed_user.title())
```

```python
responses = {}

polling_active = True

while polling_active:
    name = input('\nWhat is your name?')
    response = input('What is your favorite movie?')

    responses[name] = response

    repeat = input('Would you like to let another one respond? (yes/ not)')
    if repeat == 'not':
        polling_active = False
    
print('\n-----Poll Results-----')
for name, response in responses.items():
    print(name + "'s favorite movie is " + response + '.')
```
