```python
def get_formatted_name(first_name, last_name, middle_name = ""):
    fullname = first_name + " " + middle_name + " " + last_name
    print(fullname.title())
```

```python
get_formatted_name("l",'j')
```

```python
def build_person(first_name, last_name):
    """返回一个字典，其中包含有关一个人的信息"""
    person = {'first':first_name, "last":last_name}

    return person

musician = build_person("L","J")
print(musician)
```

```python
def greet_user(names):
    for name in names:
        print("Hello,",name.title() + "!")
```

```python
names = list(map(str,range(1,4)))
```

```python
greet_user(names)
```

```python
def show_numbers(numbers):
    for number in numbers:
        print(number)
```

```python
numbers = list(map(str,range(1,5)))
```

```python
show_numbers(numbers)
```

```python
def make_great(numbers):
    i = 0
    while i < len(numbers):
        numbers[i] += " is great."
        i += 1
    return numbers
```

```python
make_great(numbers)
```

```python
show_numbers(numbers)
```

```python
numbers2 = make_great(numbers[:])
```

```python
show_numbers(numbers)
```

```python
show_numbers(numbers2)
```

```python
def make_pizza(*toppings):
    for topping in toppings:
        print("- " + topping)
```

```python
make_pizza('1','2')
```

```python
def build_profile(name, **user_info):
    profile = {}
    profile["name"] = name

    for k, v in user_info.items():
        profile[k] = v
    
    return profile
```

```python
build_profile("lihengjin",city="SMX", gender="male")
```

```python
def material(*materials):
    print("The material of your pizza is following: ")
    for material in materials:
        print('- ' + material)
```

```python
pizzas = list(map(str,range(1,5)))
material(*pizzas)
```

```python
def make_pizza(size, *toppings):
    print('\nMaking a ' + str(size) +
    "-inch pizza with the: ")
    for topping in toppings:
        print("- " + topping)
```

```python
make_pizza(3, "1", "2")
```
